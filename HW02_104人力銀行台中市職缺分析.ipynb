## 題目

104人力銀行台中市職缺分析

## 摘要

本報告旨在透過網路爬蟲技術，收集104人力銀行網站上的台中市工作職缺，為在台中尋找打工機會的求職者提供參考。

## 引言

隨著職場競爭的激烈化，求職者對於職缺資訊的需求日益增加。104 求職網作為台灣知名的求職平台，提供豐富的職缺選擇。
本次爬蟲的目標是獲取 104 求職網上台中市的職缺資訊，分析其薪資、公司地點、公司規模等資訊，為求職者提供決策支持。

## 方法

- 目標網站描述
  - 目標網站：104人力銀行 台中市職缺頁面。  
 https://www.104.com.tw/jobs/search/?area=6001008000&jobsource=index_s&mode=s&page=2
  - 頁面結構：包含職務名稱、公司名稱、薪水、地區等資訊。

- 工具與技術
  - 使用 Python 的 `requests` 庫發送 HTTP 請求，獲取 104 職缺的 AJAX 資料。
  - 使用 Pandas 函式庫處理和組織爬取到的職缺資料，例如合併不同頁面的資料和排序。
  - 使用 Google Colab 的認證功能連接 Google Sheets API，並使用 `gspread` 函式庫將資料寫入 Google Sheet。


```python
import requests
import pandas as pd
from IPython.display import clear_output

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/134.0.0.0 Safari/537.36',
           'Referer': 'https://www.104.com.tw/'}
url= 'https://www.104.com.tw/jobs/search/list?ro=1&indcat=1003000000&area=6001008000&order=11&asc=0&page=&mode=l'
resp = requests.get(url, headers=headers)
pd.DataFrame(resp.json()['data']['list']).head(3)
```

```python
headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/134.0.0.0 Safari/537.36',
         'Referer': 'https://www.104.com.tw/'}

df = []

for page in range(1,10):
    url= f'https://www.104.com.tw/jobs/search/list?ro=1&indcat=1003000000&area=6001008000&order=11&asc=0&page={page}&mode=l'
    resp = requests.get(url, headers=headers)
    print(url)

    ndf = pd.DataFrame(resp.json()['data']['list'])
    df.append(ndf)
    if ndf.shape[0] < 30:
        break

df = pd.concat(df, ignore_index=True)
df
```

## 問題與挑戰

- 技術挑戰
